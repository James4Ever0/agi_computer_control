```python
verbal_q_star_critic = """
**You choose the action (option) based on the estimated Q function of the action.**
**You estimated the Q value of actions, assuming a scale of 0 to 10.**
In the context of reinforcement learning, the Q-function, often represented as
Q(s,a), is a fundamental concept used in Q-learning, which is a model-free, off-policy algorithm.
Definition and Purpose
Q-function (Quality Function): The Q-function estimates the quality of a state-action combination,
essentially representing the expected cumulative reward for taking an action a in a given state s and then following a certain policy thereafter.
Goal: The primary goal in Q-learning is to learn a policy that tells an agent what action to take under what circumstances.
This policy is derived by finding the action that maximizes the Q-value for each state.
Components of the Q-Function
State s: Represents the current situation or environment the agent is in.
Action a: Represents the possible action the agent can take in the given state.
Reward r: Immediate reward received after taking action a in state s.
Discount Factor γ: A value between 0 and 1 that reduces the weight of future rewards, reflecting the preference for immediate rewards over distant rewards.

Q-Learning Algorithm
Updating Rule: The Q-values are updated iteratively using the following formula:

Q(s,a)←Q(s,a)+α(r+γmax a′ Q(s′,a′)−Q(s,a))

Here,
α is the learning rate, s′is the new state after taking action a, and a′are the possible actions in the new state.

Convergence: Over time and through repeated updates,
the Q-values converge to optimal values that represent the expected sum of future rewards for each action in each state.

Role in Decision Making
The learned Q-function is used by the agent to make decisions: in any given state s,
the agent chooses the action a that maximizes the Q-value, Q(s,a).
"""
```

```python
from colorama import Fore

from camel.configs import ChatGPTConfig
from camel.societies import RolePlaying
from camel.utils import print_text_animated
from camel.types import ModelType

task_prompt = "Develop a theoretical framework or algorithm that could potentially solve the P versus NP problem"
model_config = ChatGPTConfig(temperature=0.8, n=3)
assistant_agent_kwargs = dict(model_config=model_config)
user_agent_kwargs = dict(model_config=model_config)
critic_kwargs = dict(verbose=True)


print(Fore.YELLOW + f"Original task prompt:\n{task_prompt}\n")
role_play_session = RolePlaying(assistant_role_name="AGI",
                                user_role_name="Computer Scientist",
                                critic_role_name="Q*",
                                task_prompt=task_prompt,
                                with_task_specify=False,
                                with_critic_in_the_loop=True,
                                critic_criteria=verbal_q_star_critic,
                                assistant_agent_kwargs=assistant_agent_kwargs,
                                user_agent_kwargs=user_agent_kwargs,
                                critic_kwargs=critic_kwargs,
                                model_type=ModelType.GPT_4_TURBO)
print(Fore.CYAN + f"Specified task prompt:\n{role_play_session.task_prompt}\n")

print(
        Fore.GREEN +
        f"AI Assistant sys message:\n{role_play_session.assistant_sys_msg}\n")
print(Fore.BLUE +
          f"AI User sys message:\n{role_play_session.user_sys_msg}\n")
print(Fore.MAGENTA +
          f"Critic sys message:\n{role_play_session.critic_sys_msg}\n")
print(Fore.RED + f"Final task prompt:\n{role_play_session.task_prompt}\n")

chat_turn_limit, n = 50, 0
input_assistant_msg, _ = role_play_session.init_chat()
while n < chat_turn_limit:
    n += 1
    assistant_response, user_response = role_play_session.step(input_assistant_msg)
    print_text_animated(Fore.BLUE +f"AI User:\n\n{user_response.msg.content}\n")
    print_text_animated(Fore.GREEN + "AI Assistant:\n\n"f"{assistant_response.msg.content}\n")

    if "CAMEL_TASK_DONE" in user_response.msg.content:
        break

    input_assistant_msg = assistant_response.msg
```
